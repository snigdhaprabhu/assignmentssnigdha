```python
import glob

def readfiles(a):
    data_list = []
    global filescontent
    files_list = glob.glob("/Users/snigdha/Desktop/repos/notebooks/programming-for-analytics-course-material/data/"+a+"*")
    d = {}
    for filena in files_list:
            filescontent = open(filena,'r') 
            #print filena
            header = filescontent.readline()
            data_list = filescontent.readlines()
            #print data_list
            d.update(dict([(i[0], tuple(i[1:])) for i in [list(line.strip().split(',')) for line in data_list]]))

      
    filescontent.close    
    return(d)  

```

```python
def writeDict(dict, name, sep):
    if name.startswith('aapl'):
        Str = name+"-dec80-oct15"
    else:
        Str = name+"-mar86-oct15"
    with open("/Users/snigdha/Desktop/repos/notebooks/programming-for-analytics-course-material/data/"+Str+".csv", "w") as csvfile:
        csvfile.write("Date,Open,High,Low,Close,Volume,AdjClose"+ "\n")
        for i in dict.keys():            
            csvfile.write(i + "," + sep.join([str(x) for x in dict[i]]) + "\n")
        
writeDict(readfiles("aapl"),"aapl",",")     
```
